```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
V0 = 10
a = 1
m = 1/2
hbar = 1
```

```python
Ebmesh = np.arange(0.01, V0-0.01, 0.1)

def eigen_even(eb):
    return np.sqrt(V0 - eb)*np.sin(np.sqrt(V0-eb)) - np.sqrt(eb)*np.cos(np.sqrt(V0 - eb))


def eigen_odd(eb):
    return np.sqrt(V0 - eb)*np.cos(np.sqrt(V0 - eb)) - np.sqrt(eb)*np.sin(np.sqrt(V0 - eb))


def d_even(eb):
    return -1/2*np.sin(np.sqrt(V0 - eb))*(1 + np.sqrt(eb))/np.sqrt(V0-eb) - 1/2*np.cos(np.sqrt(V0-eb))*(1+1/np.sqrt(eb))

def d_odd(eb):
    return 0.5*np.cos(np.sqrt(V0 - eb))/np.sqrt(V0 - eb)*(np.sqrt(eb) - 1) + 0.5*np.sin(np.sqrt(V0-eb))*(1 - 1/np.sqrt(eb))
```

```python
plt.plot(Ebmesh, eigen_even(Ebmesh))
#plt.plot(Ebmesh, d_even(Ebmesh))
plt.axhline(y = 0, ls='--', color='black')
plt.xlabel(r'$E_b$')
plt.ylabel(r'$f(E_b)$')
```

```python
print("Value of function at Eb = 0 is negative so there is an extra state.")
eigen_even(0)
```
