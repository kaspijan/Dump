```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
Nmesh = [2, 5, 10, 20, 40, 80, 160, 300, 500, 700, 1000, 5000, 10000, 100000]
error_trap = [2.0747041268*10**-2,
3.3311132254*10**-3,
8.3319447750*10**-4,
2.0832465329*10**-4,
5.2082790806*10**-5,
1.3020799426*10**-5,
3.2552062155*10**-6,
9.2592575601*10**-7,
3.3333331335*10**-7,
1.7006802183*10**-7,
8.3333334879*10**-8,
3.3333380600*10**-9,
8.3333763252*10**-10,
8.3382572134*10**-12
]
error_simp = [
3.3715273476*10**-4,
4.2656563806*10**-2,
5.5489487462*10**-7,
3.4711891843*10**-8,
2.1699780908*10**-9,
1.3563213798*10**-10,
8.4784137012*10**-12,
6.8813674065*10**-13,
9.0627503362*10**-14,
2.4413222805*10**-14,
6.6741184647*10**-15,
1.0538081786*10**-15,
3.3370592323*10**-15,
4.0395980181*10**-15
]

error_trap=np.array(error_trap)
error_simp=np.array(error_simp)
Nmesh = np.array(Nmesh)
Nlog = np.arange(np.log10(2), np.log10(1000000))
```

```python
def func(x, a, b):
    return a*x+b

arg, cov = curve_fit(func, np.log10(Nmesh), np.log10(error_trap))
a=arg[0]
b = arg[1]
```

```python
plt.scatter(np.log10(Nmesh), np.log10(error_trap))
plt.xlabel(r'$log_{10} (N)$')
plt.ylabel(r'$log_{10}(error)$')
plt.plot(Nlog, a*Nlog+b, ls='--', color='orange', label='linear fit')
plt.legend()
```
