```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''Goal of the exercise is to approximate a linear function with Fourier series and plot the approximation for 
varying number of terms in the sum.'''
''
```

```python
xs = np.linspace(0, 2*np.pi, 1000)
ys = [np.pi/2 - x/2 for x in xs]

plt.plot(xs, ys)
plt.title('Funkcija na jednom periodu')
```

```python
##for one period

def f(x, n):
    sum=0
    for i in range(1, n+1):
        sum += np.sin(i*x)/i
    return sum

ys2 = []
for x in xs:
    ys2.append(f(x, 1))
    
plt.plot(xs,ys2)
plt.plot(xs, ys)
plt.title('1-term sum')
plt.show()


ys2 = []
for x in xs:
    ys2.append(f(x, 5))
    
plt.plot(xs,ys2)
plt.plot(xs, ys)
plt.title('5-term sum')
plt.show()


ys2 = []
for x in xs:
    ys2.append(f(x, 15))
    
plt.plot(xs,ys2)
plt.plot(xs, ys)
plt.title('15-term sum')
plt.show()
```

```python
#for multiple periods

xs = np.linspace(-2*np.pi, 6*np.pi, 1000)
def f(x, n):
    sum=0
    for i in range(1, n+1):
        sum += np.sin(i*x)/i
    return sum

ys2 = []
for x in xs:
    ys2.append(f(x, 1))
    
plt.plot(xs,ys2)
#plt.plot(xs, ys)
plt.title('1-term sum')
plt.show()


ys2 = []
for x in xs:
    ys2.append(f(x, 5))
    
plt.plot(xs,ys2)
#plt.plot(xs, ys)
plt.title('5-term sum')
plt.show()


ys2 = []
for x in xs:
    ys2.append(f(x, 15))
    
plt.plot(xs,ys2)
#plt.plot(xs, ys)
plt.title('15-term sum')
plt.show()
```

```python
xs = np.linspace(0, 2*np.pi, 1000)
ys = [np.pi/2 - x/2 for x in xs]
xs2 = np.linspace(2*np.pi, 4*np.pi, 1000)
ys2=[np.pi/2 -x/2 for x in xs]
xs3 = np.linspace(4*np.pi, 6*np.pi, 1000)
ys3=[np.pi/2 -x/2 for x in xs]
plt.plot(xs,ys, color='blue',ls ='--', label='stvarna fja')
plt.plot(xs2,ys2, color='blue',ls ='--')
plt.plot(xs3,ys3, color='blue',ls ='--')

xsf=np.linspace(0,6*np.pi, 3000)
ysf = []
for x in xsf:
    ysf.append(f(x, 1))
    
plt.plot(xsf,ysf, color='orange', label='suma')
plt.legend()
plt.title("suma $s_{1}(x)$")
plt.savefig('suma1')
```
